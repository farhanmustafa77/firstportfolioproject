```python
####### data cleansing ########
#removing non numerical values & replacing missing values
import re #a library for regular expressions
```

```python
#for the sake of visualizing only ASUS, ACER. & DELL brands. We will create a copy df for only these three 
#brands
```

```python
brands = {
    'ASUS': ['asus', 'rog'],
    'ACER': ['acer'],
    'DELL': ['dell']
}

# Function to map values to brands
def map_brand(value):
    for brand, keywords in brands.items():
        if any(keyword in value.lower() for keyword in keywords):
            return brand
    return None

# Apply the function to create a new 'brand' column
df['brand'] = df['title'].apply(map_brand)
```
