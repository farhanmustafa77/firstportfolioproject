```python
####### data cleansing ########
#removing non numerical values & replacing missing values
import re #a library for regular expressions
```

```python
df["price"] = df["price"].apply(lambda x : re.sub(r'[^0-9]', "", str(x))) #removes all non numeric in column "price"
```

```python
df["price"]
```

```python
df["price"] = df["price"].astype('int64') #change dtype object to int64 or int --> *.astype(int)
```

```python
df.dtypes
```

```python
df["discount"] = df["discount"].apply(lambda x : re.sub(r'[^0-9]', "", str(x)))
# df["RAM"] = df["RAM"].apply(lambda x : re.sub(r'[^0-9]', "", str(x)))
df["SSD"] = df["SSD"].apply(lambda x : re.sub(r'[^0-9]', "", str(x)))
# df["Display"] = df["Display"].apply(lambda x : re.sub(r'[^0-9]', "", str(x)))
```

```python
df = df.replace("", np.nan) #convert all missing values with NaN
df = df.fillna(0) #convert NaN to 0
```

```python
df["discount"] = df["discount"].astype('int64')
# df["RAM"] = df["RAM"].astype('float64')
df["SSD"] = df["SSD"].astype('int64')
# df["Display"] = df["Display"].astype('float64')
```

```python
df.rename(columns = {"SSD" : "SSD_GB"}, inplace = True)
df.rename(columns = {"price" : "price_$"}, inplace = True)
df.rename(columns = {"discount" : "discount_%_off"}, inplace = True)
```
